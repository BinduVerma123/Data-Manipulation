```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
iris=sns.load_dataset('iris')
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
```

```python
x=x.map_diag(plt.scatter)
```

```python
x=x.map_offdiag(plt.scatter)
```

```python
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist,histtype='step')
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist,histtype='step',linewidth=4)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```
